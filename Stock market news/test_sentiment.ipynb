```python
import os
import numpy as np
import pandas as pd
from google.colab import drive #เชื่อม Google drive
drive.mount('/content/drive')
path = '/content/drive/My Drive/master project' #เข้า drive
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
News_GOOGL = pd.read_csv(os.path.join(path,'News_GOOGL.csv'))
```

```python
News_GOOGL
```

```python
News_GOOGL=News_GOOGL.astype(str)
```

```python
sia = SentimentIntensityAnalyzer()
# Analyze sentiment for each summary

pop = str(News_GOOGL['summary']).encode("utf-8")
News_GOOGL['sentiment_score'] = News_GOOGL['summary'].apply(lambda pop: sia.polarity_scores(pop)['compound'])
```

```python
News_GOOGL
```

```python
sia = SentimentIntensityAnalyzer()

# วิเคราะห์ความรู้สึกของแต่ละข่าว
sentiments = []
for article in News_GOOGL:
    summary = article.get('summary', '')
    if summary:
        sentiment = sia.polarity_scores(summary)
        sentiments.append(sentiment['compound'])
```
