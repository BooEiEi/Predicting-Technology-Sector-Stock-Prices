#Merge

```python
from functools import reduce

def merge_multiple_dataframes(dataframes, on, how='left'):

    if not dataframes:
        return None  # Return None if the list is empty

    merged_df = reduce(lambda left, right: pd.merge(left, right, on=on, how=how), dataframes)
    return merged_df
```

```python
merged_df = merge_multiple_dataframes([Stockprice_META, Stockprice_AAPL, Stockprice_NVDA, Stockprice_MSFT, Stockprice_GOOGL, daily_avg_sentiment_Nozero_META,
                                       daily_avg_sentiment_Nozero_AAPL, daily_avg_sentiment_Nozero_NVDA, daily_avg_sentiment_Nozero_MSFT, daily_avg_sentiment_Nozero_GOOG,
                                       FED, UNEMPLOYMENT, CPI, US_Treasury_Yield, RETAIL_SALES, US_inflation_rate, Real_GDP, US_Dollar_Index, SP500, NASDAQ, WTI, BTC, GOLD], on='date', how='left')

# แสดงผลลัพธ์
print(merged_df)
```
